```python
import csv
import numpy as np
import pandas as pd
pd.set_option('display.min_rows',20)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cs = open('company_sales_data.csv')
companyDF = pd.read_csv(cs)
companyDF.head(12)
```

```python
#exercício 1
companyDF.plot(kind="line",title="Company profit per month", x="month_number",y="total_profit")
plt.axis([0,12,100000,500000])
profitList=companyDF['total_profit'].tolist()
monthList=companyDF['month_number'].tolist()
```

```python
df = pd.read_csv("company_sales_data.csv")
profitList = df ['total_profit'].tolist()
monthList  = df ['month_number'].tolist()

plt.plot(monthList, profitList, label = 'Lucro ultimo ano', 
      color='r', marker='o', markerfacecolor='k', 
      linestyle='--', linewidth=3)
      
plt.xlabel('Meses do ano')
plt.ylabel('Numero de unidades vendidas')
plt.legend(loc='lower right')
plt.title('Vendas ultimo ano')
plt.xticks(monthList)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

```python
df = pd.read_csv("company_sales_data.csv")
monthList  = df ['month_number'].tolist()
Cremefacial   = df ['facecream'].tolist()
LimpezaFacial = df ['facewash'].tolist()
PastaDentes = df ['toothpaste'].tolist()
Sabonete = df ['bathingsoap'].tolist()
Shampoo = df ['shampoo'].tolist()
Hidratante = df ['moisturizer'].tolist()

plt.plot(monthList, Cremefacial,   label = 'Vendas Creme Facial', marker='o', linewidth=3)
plt.plot(monthList, LimpezaFacial,   label = 'Vendas Limpeza Facial',  marker='o', linewidth=3)
plt.plot(monthList, PastaDentes, label = 'Vendas Pasta de Dentes', marker='o', linewidth=3)
plt.plot(monthList, Sabonete, label = 'Vendas Sabonetes', marker='o', linewidth=3)
plt.plot(monthList, Shampoo, label = 'Vendas Shampoo', marker='o', linewidth=3)
plt.plot(monthList, Hidratante, label = 'Vendas hidratante', marker='o', linewidth=3)

plt.xlabel('Meses do Ano')
plt.ylabel('Unidade de vendas')
plt.legend(loc='upper left')
plt.xticks(monthList)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Vendas mensais')
plt.show()

```

```python
df = pd.read_csv("company_sales_data.csv")
monthList  = df ['month_number'].tolist()
labels = ['Creme Facial', 'Limpeza Facial', 'Pasta de dentes', 'sabonete', 'Shampoo', 'Hidratante']
salesData   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]

plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Total de vendas Anual')
plt.show()

```
